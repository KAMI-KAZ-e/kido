```python
import matplotlib.pyplot as plt
import os
import seaborn as sns
from scipy.io import wavfile
from pydub import AudioSegment
sns.set()

def convert_mp4_to_wav(file, plot_show = False, plot_save = False):
    """
    Reads a mp4 (=.m4a) file and converts it to .wav format.
    It also plots the wav file and saves the figure if desired.
    
    file = file name with complete path (must have .m4a extension)
    plot_show = show the plot? (plot_show = False, if not specified)
    plot_save = save the plot? (plot_save = False, if not specified)

    While saving, it keeps the file name and just changes the extension to .wav.
    For example, it will save '/Downloads/example.m4a' as '/Downloads/example.wav'

    """
    
    # Create the wav file name with path
    file_wav = file[:-4] + ".wav"

    # Convert mp4(=m4a) to wav (need to install pydub and ffmpeg)
    # https://pythonbasics.org/convert-mp3-to-wav/
    # https://www.youtube.com/watch?v=8nbuqYw2OCw
    # https://github.com/jiaaro/pydub
    audio = AudioSegment.from_file(file, "mp4")
    audio.export(file_wav, "wav")

    # Read the wav file
    sample_rate, samples = wavfile.read(file_wav)
    samples1D = samples[:,0]
    
    # Read and plot the wav
    fig, ax = plt.subplots(1, figsize=(10,6))
    ax.plot(samples1D)
    ax.set(xlabel="Sample #", ylabel="Amplitude", title=file_wav)
    if plot_save: fig.savefig(file_wav[:-4] + '.png', dpi = 300, transparent=True)
    if not plot_show: plt.close(fig)
    plt.show()
```
