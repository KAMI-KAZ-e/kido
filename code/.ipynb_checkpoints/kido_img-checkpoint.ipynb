```python
import matplotlib
import matplotlib.pyplot as plt
from scipy.io import wavfile
import numpy as np
import os

# https://stackoverflow.com/questions/47147146/
# save-an-image-only-content-without-axes-or-anything-else-to-a-file-using-matl
# https://stackoverflow.com/questions/34162443/
# why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python
def graph_spectrogram(input_path, wav_file, output_path, output_fname, close_fig = True):
    """
    Converts wav to spectrogram and saves as png.
    
    
    """
    rate, data = wavfile.read(input_path + wav_file)
    data1D = data[:, 0]
    fig, ax = plt.subplots(1)
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    pxx, freqs, bins, im = ax.specgram(x=data1D, Fs=rate, noverlap=384, NFFT=512)
    ax.axis('off')
    fig.savefig(output_path + output_fname, dpi=300, frameon='false')
    if close_fig:
        matplotlib.pyplot.close(fig)
```

```python
list_dirs = [
    #"batch00_audioSplit", 
    #"batch01_audioSplit__dBFS", 
    #"batch02_audioSplit__dBFSm10",
    
    #"batch50_audioSplit__dBFSm10",
    
    #"batch80_audioSplit__dBFSm10", 
    #"batch81_audioSplit__dBFSm10", 
    #"batch82_audioSplit__dBFSm10",
    
    #"batch90_audioSplit__dBFSm10",
    #"batch91_audioSplit__dBFSm10",
    #"batch92_audioSplit__dBFSm10"
]

for dir_ in list_dirs:
    path_project = "Documents/Learning/Projects/kido/"
    path_wav = path_project + dir_ + "/"
    path_img_base = path_project + "images/train/"
    list_files = os.listdir(path_wav)
    list_wav = [file for file in list_files if file.endswith('.wav')]
    print("\n" + dir_, len(list_wav))
    for file_ in list_wav:
        fname = file_[:-4]
        file_png = fname + '.png'
        path_img = path_img_base + fname.split("_")[-1] + "/"
        graph_spectrogram(path_wav, file_, path_img, file_png)
        print('.', end='')
```

```python
import random
import shutil

# Train-dev split
# https://cs230-stanford.github.io/train-dev-test-split.html
path_project = "Documents/Learning/Projects/kido/"
path_train = path_project + "images/train/"
path_dev = path_project + "images/dev/"

# Loop over a-to-z
n_alpha = 26
for i in range(n_alpha):
    alpha = chr(i + ord('a'))
    dir_train = path_train + alpha + "/"
    dir_dev = path_dev + alpha + "/"
    
    # If we don't already have 100 dev images per class
    list_dev = os.listdir(dir_dev)
    list_dev = [file for file in list_dev if file.endswith('.png')]
    if len(list_dev) < 100:
        print(i, dir_train, dir_dev)
        n_move = 100 - len(list_dev)
        
        # Get all the training file names
        list_files = os.listdir(dir_train)
        list_png = [file for file in list_files if file.endswith('.png')]
        random.shuffle(list_png)
        random.shuffle(list_png)
        list_dev = list_png[-n_move:]

        # Move the dev images
        # https://stackabuse.com/how-to-create-move-and-delete-files-in-python/
        # Use shutil.move(source_file_with_path, destination_file_with_path)
        for file in list_dev:
            shutil.move(dir_train + file, dir_dev + file)
            
# Plot stats for verification
stats_train = [0] * 26
stats_dev = [0] * 26
for i in range(n_alpha):
    alpha = chr(i + ord('a'))
    dir_train = path_train + alpha + "/"
    dir_dev = path_dev + alpha + "/"
        
    # Count the number of train files
    list_files = os.listdir(dir_train)
    list_png = [file for file in list_files if file.endswith('.png')]
    stats_train[i] = len(list_png)
    
    # Count the number of dev files
    list_files = os.listdir(dir_dev)
    list_png = [file for file in list_files if file.endswith('.png')]
    stats_dev[i] = len(list_png)
    
for i in range(26):
    alpha = chr(i + ord('a'))
    print(i, alpha, stats_train[i], stats_dev[i])

```

```python
# Bar chart for train-dev split

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

path_figures = path_project + "figures/"

bottom = stats_train
labels = [chr(i + ord('a')) for i in range(len(stats_train))]
x = [i for i in range(len(stats_train))]
plt.bar(x, stats_train, align='center', alpha=0.8)
plt.bar(x, stats_dev, bottom = bottom, align='center', alpha=0.8)
plt.xticks(x, labels)
plt.ylabel('# of samples')
#plt.title('Train-Dev Split')
plt.legend(['Train', 'Dev'], loc=0)
#plt.savefig(path_figures + 'train_dev_split.png', dpi = 1000)
plt.show()


```
